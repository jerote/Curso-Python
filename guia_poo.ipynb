```python
# Ejercicio 1

# Vamos a crear una clase llamada Persona. Sus atributos son: nombre, edad y DNI.
# Construir los siguientes mÃ©todos para la clase:
# â— Un constructor, donde los datos pueden estar vacÃ­os.
# â— mostrar(): Muestra los datos de la persona.
# â— esMayorDeEdad(): Devuelve un valor lÃ³gico indicando si es mayor de edad.

class Persona: # creo la clase Persona. no lleva parÃ©ntesis
    
    def __init__(self, nombre="sin nombre", edad=0, dni=0): # constructor incializador de Persona
        self.nombre = nombre
        self.edad = edad
        self.dni = dni
        
    def mostrar(self): # mÃ©todo muestra datos de Persona
        print(f"Nombre: {self.nombre.title()}\nEDAD: {self.edad} aÃ±os.\nDNI: {self.dni}\n")
    
    def esMayorDeEdad(self): # mÃ©todo verifica si Persona es mayor o menor de edad.
        if self.edad >= 18:
            return True
        else:
            return False
        
jero = Persona("Jeronimo", 41, 28507346)
fran = Persona("Francisco", 39, 30654564)
juan = Persona("Juan Cruz", 19, 32987987)

jero.mostrar()
fran.mostrar()
juan.mostrar()
```

```python
# Ejercicio 2

# Crea una clase llamada Cuenta que tendrÃ¡ los siguientes atributos: titular (que es una
# persona) y cantidad (puede tener flotantes). Construir los siguientes mÃ©todos para la
# clase:
# â— Un constructor, donde los datos pueden estar vacÃ­os.
# â— mostrar(): Muestra los datos de la cuenta.
# â— ingresar(cantidad): se ingresa una cantidad a la cuenta, si la cantidad introducida
# es negativa o 0 (cero), no se harÃ¡ nada.
# â— retirar(cantidad): se retira una cantidad a la cuenta. La cuenta puede estar en
# nÃºmeros rojos (se puede retirar mÃ¡s de lo que se tiene).

class Cuenta: # defino la clase Cuenta
    
    def __init__(self, titular=Persona(), saldo=float(0)): # constructor con atributos titular y cantidad.
        self.titular = titular
        self.saldo = saldo

    def mostrar(self): # Muestra la info de la cuenta del titular.
        self.titular.mostrar()
        print(f"Saldo disponible: {self.saldo}")

    def ingresar(self, monto_i=0): # ingresa dinero a la cuenta.
        
        monto_i = int(input("Â¿CuÃ¡nto depositÃ¡s? $"))
        
        if monto_i > 0:
            self.saldo += monto_i
            print(f"Ingresaste: {monto_i}")
        print(f"Saldo disponible: {self.saldo}\n")
        
    def retirar(self, monto_r=0): # retira dinero de la cuenta.
        
        monto_r = int(input("Â¿CuÃ¡nto retirÃ¡s? $"))
            
        if monto_r > 0:
            self.saldo -= monto_r
            print(f"Retiraste: {monto_r}")
        print(f"Saldo disponible: {self.saldo}\n")


jero.cuenta = Cuenta(jero, 50000)

jero.cuenta.mostrar()
jero.cuenta.ingresar()
jero.cuenta.retirar()
```
